````python
# CELL 4
# Here are the functions that will compute our strategy signal. Each function takes a dataframe `df`, along with
# more optinal parameters.
# Each function has to set the "strat_signal" column in the `df`. The values will range from -1 to 1, telling us to
# go short (-1), neutral (0), or long (1). Intermediate values are also supported.

def future_info_signal(df):
    # We take a peek at tomorrow's price change to decide what we should do today.
    # Obviously don't do this for real strategies!
    df["strat_signal"] = np.sign(df["pct_change"].shift(-1).fillna(0))
    return df

def constant_signal(df, signal=1):
    df["strat_signal"] = signal
    return df

def mean_crossover_signal(df, mean_days=7):
    # Compute rolling mean of the price
    df["mean"] = df['close'].rolling(mean_days).mean()
    # If close > mean, then the sign will be +1; if close < mean, then the sign will be -1; otherwise it's 0.
    df["strat_signal"] = np.sign(df["close"] - df["mean"])
    return df
````

````python
## Note!
## This is cleaned up as `fast_breakout()` at the bottom of this section.



def breakout(df, back=1, roll=30):
  df = support_resistance(df, roll)
  pctsign = np.sign(df["pct_change"].fillna(0))
  
  #breakout up
  df["breakup"] = df["close"] > df["resistance"].shift(1).fillna(0)
  
  #I realize multi-case is kludgier than it could be. I wanted to switch
  #to a style using shift(0) for the first-case so I can better reuse code.
  #I started with that, but split off the back>=2 cases while fixing a bug,
  #and haven't reverted it back yet.
  #Besides, back=1 was by far the best performer?
  if back>1:
    for x in range(1,back):
      resist = df["resistance"].shift(x).fillna(0) > df["resistance"].shift(x+1).fillna(0)
      df["breakup"] = df["breakup"] & resist
  
  # breakout down
  df["breakdown"] = df["close"] < df["support"].shift(1).fillna(0)
  
  if back>1:
    for x in range(1,back):
      support = df["support"].shift(x).fillna(0) < df["support"].shift(x+1).fillna(0)
      df["breakdown"] = df["breakdown"] & support
  
  df["breakout"] = df["breakup"] | df["breakdown"]
  df["strat_signal"] = df["breakout"].astype(int) * pctsign
  #I realize this is a somewhat awkward way to handle this, but... backfilling so it holds a position until the next opposing signal.
  df["strat_signal"] = df["strat_signal"].replace({0:np.nan}).fillna(method="backfill")
  return df

# Fixed an error where I was | ing booleans instead of &ing; weird even-numbers behavior (excessive swaps) was a clue
# It's not holding position on the 0s. Hm... could turn all 0s into na, and backfill?
````

````python
def volume_spike_indicator(df, vroll=14, currency="usd"):
  #Note: Returning vol_sign as boolean,
  #but very seriously considered returning vol_sign as int.
  #Bool breaks a mild convention, but seemed a tiny bit better for debugging.
  #If it wasn't my own script code, though, I'd go with whatver convention was set.
  df["vol_mean"] = df["volume_{}".format(currency)].rolling(vroll).mean()
  # Easy vol spike indicator: vol_usd>vol_mean
  df["vol_sign"] = df["volume_{}".format(currency)] > df["vol_mean"]
  return df



def support_resistance_volume(df, roll=7, vroll=14, currency="usd"):
  
  newdf = breakout_indicator(df, roll)
  newdf = volume_spike_indicator(newdf, vroll, currency)
  pctsign = np.sign(df["pct_change"].fillna(0))
  
  # At this point, it has:
  # support & resistance columns,
  # the zero-ridden form of breakout signal,
  # rolling volume,
  # and boolean vol_sign indicating volume spikes
  
  # Pass the boolean vol_sign over strat_signal, then backfill
  
  # df.where replaces when the condition is False
  # df.mask replaces when the condition is True
  
  df["strat_signal"] = df["strat_signal"].mask(df["vol_sign"], np.nan)
  

  #Backfilling so it holds a position until the next opposing signal.
  df["strat_signal"] = df["strat_signal"].replace({0:np.nan}).fillna(method="backfill")
  return df
````
